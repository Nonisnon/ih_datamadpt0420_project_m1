### clean age

```python
def remove_yearsold(x):
    try:
        return int(x.replace(' years old',""))
    except:
        return x
    
data_raw['age'] = data_raw['age'].apply(remove_yearsold) 


def age_calculation(x):
    try:
        if len(str(x))>3:
            year = 2016
            age = year - x
            return age
        else:
            return x
    except:
        return x

data_raw['age'] = data_raw['age'].apply(age_calculation)
data_raw.head()
```

### clean gender pycharm

```python
def clean_gender(df):
    #df['gender'] = df['gender'].astype(str)
    df['gender'] = df['gender'].replace('male','Male')\
    .replace('FeMale','Female')\
    .replace('Fem','Female')\
    .replace('female','Female')
    return df
```

```python
def clean_columns(df):
    df.rename(columns={'title': 'Job Title', 'country_name':'Country', 'age': 'Age'}, inplace=True)
    #df.loc[df['Job Title'].isnull(),'Job Title']='Unemployed'
    df['Job Title'].fillna('Unemployed', inplace = True)
    df['normalized_job_code'].fillna('Unemployed', inplace = True)
    return df
```

```python
def analyze(df):
    grouped = df.groupby(['Country', 'Job Title', 'Age Group']).agg({'Quantity':'count'}).reset_index()
    return grouped
```
