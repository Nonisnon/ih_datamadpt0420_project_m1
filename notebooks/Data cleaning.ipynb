```python
path = '/Users/Nonis/desktop/ih_datamadpt0420_project_m1/data/raw'

data_raw = pd.read_csv(f'{path}/data_raw.csv', sep=';') 

data_raw.head()
```

### clean age

```python
def remove_yearsold(x):
    try:
        return int(x.replace(' years old',""))
    except:
        return x
    
data_raw['age'] = data_raw['age'].apply(remove_yearsold) 


def age_calculation(x):
    try:
        if len(str(x))>3:
            year = 2016
            age = year - x
            return age
        else:
            return x
    except:
        return x

data_raw['age'] = data_raw['age'].apply(age_calculation)
data_raw.head()
```

### clean gender

```python
import re
data_raw['gender'].value_counts()
    
```

```python
def clean_gender(x):
    x_lower = x.lower()
    if x_lower == 'fem':
        xf = x_lower.replace('fem','female')
        return xf
    else:
        return x_lower
    
data_raw['gender'] = data_raw['gender'].apply(clean_gender)
data_raw.head()
```
