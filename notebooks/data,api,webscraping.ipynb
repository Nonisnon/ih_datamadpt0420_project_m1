### API

```python
import requests
```

```python
jobs_api = []
job_codes = set(data_raw['normalized_job_code'])

for i in job_codes:
    response = requests.get(f'http://api.dataatwork.org/v1/jobs/{i}')
    jobs_json = response.json()
    jobs_api.append(jobs_json)
    
```

```python
jobs = pd.DataFrame(jobs_api)
jobs = jobs.rename(columns = {'uuid':'normalized_job_code'})
data_jobs_api = jobs[['normalized_job_code', 'title', 'normalized_job_title']]
data_jobs_api
```

### WEB SCRAPING

```python
import requests
from bs4 import BeautifulSoup
import lxml.html as lh
import pandas as pd
url = 'https://ec.europa.eu/eurostat/statistics-explained/index.php/Glossary:Country_codes'
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
table = soup.find('table')
df_table = pd.DataFrame(table)
```

```python
import re
list_a = []
rows = table.find_all('tr')
for tr in rows:
    element = tr.find_all('td')
    for td in element:
        list_a.append(td.text)
#print(list_a)

list_cleaned = []
for i in list_a:
    f = i.replace('\n', '').replace('(','').replace(')','')
    list_cleaned.append(f)
#print(list_cleaned)
```

```python
row_split = 2
rows_refactored = [list_cleaned[x:x+row_split] for x in range(0, len(list_cleaned), row_split)]
data_countries =pd.DataFrame(rows_refactored,columns = {'country_name','country_code'})
data_countries
```
